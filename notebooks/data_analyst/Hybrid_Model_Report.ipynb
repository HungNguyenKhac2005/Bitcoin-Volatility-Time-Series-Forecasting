```python
fig, ax = plt.subplots(figsize=(20, 5))

plt.plot(test_set_timestamp['timestamp'], preds_real_cnn,
         color='yellow')

plt.plot(test_set_timestamp['timestamp'], preds_real_gru,
         color='blue')

plt.plot(test_set_timestamp['timestamp'], preds_real_tft,
         color='green')

# Đường Hybrid (màu đỏ) đậm hơn
plt.plot(test_set_timestamp['timestamp'], preds_real_hybrid,
         color='red', linewidth=3)

plt.plot(test_set_timestamp['timestamp'], preds_real_lstm,
         color='purple', linewidth=3)
# Đường thực tế (màu đen) đậm hơn
plt.plot(test_set_timestamp['timestamp'], test_set_df['volatility_14_annual'],
         color='black', linewidth=3)

ticks = np.linspace(0, len(test_set_timestamp['timestamp']) - 1, 10, dtype=int)
plt.xticks(ticks, rotation=45)
fig.suptitle('Actual vs Predicted Bitcoin Volatility',fontweight=2.5,fontsize=16)
ax.set_xlabel('Timestamp',fontweight=1.5,fontsize=16)
ax.set_ylabel('Volatility',fontweight=1.5,fontsize=16)
plt.show()
```

```python
mae, rse, rmse, mape = {},{},{},{}
preds = [preds_real_cnn, preds_real_gru, preds_real_hybrid, preds_real_lstm, preds_real_tft]
name_model = ['cnn','gru','hybrid','lstm','tft']

for i in range(len(name_model)):
    mae_temp = mean_absolute_error(test_set_df['volatility_14_annual'], preds[i])
    mae[name_model[i]] = mae_temp
    rse_temp = mean_squared_error(test_set_df['volatility_14_annual'], preds[i])
    rse[name_model[i]] = rse_temp
    rmse_temp = np.sqrt(mean_squared_error(test_set_df['volatility_14_annual'], preds[i]))
    rmse[name_model[i]] = rmse_temp
    mape_temp = mean_absolute_percentage_error(test_set_df['volatility_14_annual'], preds[i])
    mape[name_model[i]] = mape_temp
print(mae)
print(rse)
print(rmse)
print(mape)

```

```python
df_metrics = pd.DataFrame({
    'MAE': mae,
    'MSE': rse,
    'RMSE': rmse,
    'MAPE': mape
})
df_metrics = df_metrics.reset_index().rename(columns={'index': 'Model'})
df_metrics
```

```python
df_metrics_melt = pd.melt(df_metrics,
                           id_vars='Model',
                           value_vars=['MAE',"MSE","RMSE","MAPE"],
                           var_name='Metric',
                           value_name='Values')
df_metrics_melt
```

```python
palette = {
    'cnn': 'yellow',
    'gru': 'blue',
    'hybrid': 'red',
    'lstm': 'purple',
    'tft': 'green'
}
plt.figure(figsize=(15,5))
sns.barplot(x=df_metrics_melt['Metric'],y=df_metrics_melt['Values'],hue=df_metrics_melt['Model'],palette=palette,alpha=0.75);
plt.title('Comparison of MAE, MSE, RMSE, and MAPE Across Models',fontweight=1, fontsize=16, pad=15);
```

```python
mape_plot = df_metrics['MAPE']*100
plt.figure(figsize=(15, 5))
plt.plot(name_model, mape_plot, marker='o', linestyle='-', color='tab:red')

for x, y in zip(name_model, mape_plot):
    plt.text(x, y, f'{y:.4f}' + '%', ha='center', va='bottom', fontsize=10)

plt.title('Mean Absolute Percentage Error (MAPE) Comparison Across Models')
plt.xlabel('Model')
plt.ylabel('MAPE')
plt.grid(alpha=0.3)
plt.show()
```
