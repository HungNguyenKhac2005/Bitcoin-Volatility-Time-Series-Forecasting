### Import thư viện 📦

```python

# Import các thư viện cần thiết
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.stattools import adfuller
import warnings
warnings.filterwarnings("ignore")
```

### Định nghĩa hàm sử dụng ⚙️

```python

# Load dữ liệu đã xử lý từ bước preprocessing
df = pd.read_csv(r"C:\học Python\Data Science and Atifficial Interligent\Các project data science và AI\Data Analsys\Phân tích dữ liệu và xây dựng hybrid model trong việc dự đoán volatility của Bitcoin tại thi trường crypto toàn cầu\data\Bitcoin_time_series_processed.csv")
df.drop(columns='Unnamed: 0',inplace=True)
df
```

```python

# tính log_return dựa trên giá, bởi vì model garch chỉ học với dữ liệu đầu vào là log_return
df['log_return'] = np.log(df['close'] / df['close'].shift(1))
df = df.dropna()
df
```

```python

# scaler dữ liệu
df['log_return'] = df['log_return'] * 100
df
```

```python

# Kiểm định stationary
result = adfuller(df['log_return'])
print("ADF Statistic:", result[0])
print("p-value:", result[1])
```

```python

# lấy ra những cột cần thiết để training model
bitcoin_garch_dataset = pd.DataFrame()
bitcoin_garch_dataset['timestamp'] = df['timestamp']
bitcoin_garch_dataset['volatility_14_annual'] = df['volatility_14_annual']
bitcoin_garch_dataset['log_return'] = df['log_return']
bitcoin_garch_dataset
```

```python

# Xuất dữ liệu sang file .csv
bitcoin_garch_dataset.to_csv("bitcoin_garch_dataset.csv")
```
