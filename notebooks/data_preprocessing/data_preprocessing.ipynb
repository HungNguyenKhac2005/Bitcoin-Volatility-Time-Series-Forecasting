```python

# Lấy ra các cột có giá trị null
null_columns = df.columns[df.isnull().any()].to_list()
null_columns
```

```python

# lấy ra tổng null của các cột
df_totla_null = df.isnull().sum().reset_index()
df_totla_null.rename(columns={0:'total null'},inplace=True)
df_totla_null
```

```python

# xóa đi các giá trị null
df = df.dropna()
```

```python

# Lấy ra các cột có giá trị null
null_columns = df.columns[df.isnull().any()].to_list()
null_columns
```

```python

# kiểm tra các giá trị trùng lặp
total_duplicate = df.duplicated().sum()
total_duplicate
```
